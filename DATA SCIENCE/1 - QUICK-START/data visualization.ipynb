```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=x**2 + 3
print(y)
```

```python
#to see available styles for graph printing
plt.style.available
```

```python
plt.style.use('seaborn')  #font for graph printing
```

```python
#plotting x on x-axis and y on y-axis
plt.plot(x,y,label='stock price') #to execute this u need to call legend()
plt.plot(x,x**2,label='stock price 2')
plt.xlabel('time stamp') #x-axis name
plt.ylabel('price')     #y-axis name
plt.title('my line plot')
plt.legend()
plt.show()
```

```python
#scatter plotting i.e points
plt.scatter(x,y,label='stock price') #to execute this u need to call legend()
plt.scatter(x,x**2,label='stock price 2')
plt.xlabel('time stamp') #x-axis name
plt.ylabel('price')     #y-axis name
plt.title('my scatter plot')
plt.legend()
plt.show()
```
