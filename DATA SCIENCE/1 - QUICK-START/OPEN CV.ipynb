```python
#to print image as RGB image i.e true image
img=cv2.imread('D:\waterfall.jpg')
rgb_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert colour

plt.imshow(rgb_image)

```

```python
#to remove axis
img=cv2.imread('D:\waterfall.jpg')
rgb_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert colour
plt.axis('off') #to remove axis
plt.imshow(rgb_image)
plt.show()
```

```python
#crop image
crop_image=rgb_image[200:800,500:1250,:] #[no.rows startpixel:endpixel,column:column,channel:channel]
plt.imshow(crop_image)                                                           
plt.show()
```

```python
#dimension of cropped image
a=crop_image.shape
print(a)
pixels=a[0]*a[1] #no. of total pixels
print(pixels)
```

```python
#how to resize image resolution
crop_image=cv2.resize(crop_image,(100,100))# 100 cross 100 pixels
plt.imshow(crop_image)
plt.title('nature'+str(crop_image.shape))# str will give shape i.e pixel values
```
